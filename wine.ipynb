```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('winequalityN.csv')
data.head()
```

```python
data.hist(bins=25, figsize=(10,10))
plt.show()
```

```python
plt.figure(figsize=[10, 6])
plt.bar(data['quality'], data['alcohol'], color='red')
plt.xlabel('quality')
plt.ylabel('alcohol')
```

```python
plt.figure(figsize=[19, 10], facecolor='white')
sns.heatmap(data.corr(), annot=True)
```

```python
for a in range(len(data.corr().columns)):
    for b in range(a):
        if abs(data.corr().iloc[a,b]) >0.7:
            name = data.corr().columns[a]
            print(name)
```

```python
new_df=data.drop('total sulfur dioxide',axis=1)
```
